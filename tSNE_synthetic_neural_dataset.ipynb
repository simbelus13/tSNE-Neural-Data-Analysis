```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Simulate synthetic neural activity
np.random.seed(42)
n_neurons = 50   # Number of neurons
n_time_steps = 1000  # Number of time steps
neural_data = np.random.rand(n_time_steps, n_neurons)  # Random activity (0 to 1)

# Add structured patterns (simulate neural states)
for i in range(0, n_time_steps, 200):
    neural_data[i:i+100, :] += np.sin(np.linspace(0, np.pi, n_neurons))

# Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
data_2d = tsne.fit_transform(neural_data)

```

```python
for perplexity in [5, 30, 50]:
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    data_2d = tsne.fit_transform(neural_data)

    plt.figure(figsize=(8, 6))
    plt.scatter(data_2d[:, 0], data_2d[:, 1], c=np.arange(len(data_2d)), cmap=plt.cm.Spectral)
    plt.title(f"t-SNE with Perplexity={perplexity}")
    plt.xlabel("t-SNE Dimension 1")
    plt.ylabel("t-SNE Dimension 2")
    plt.colorbar(label="Time Index")
    plt.show()
```

```python
for lr in [10, 50, 200, 500, 750, 1000]:
    tsne = TSNE(n_components=2, learning_rate=lr, random_state=42)
    data_2d = tsne.fit_transform(neural_data)

    plt.figure(figsize=(8, 6))
    plt.scatter(data_2d[:, 0], data_2d[:, 1], c=np.arange(len(data_2d)), cmap=plt.cm.Spectral)
    plt.title(f"t-SNE with Learning Rate={lr}")
    plt.xlabel("t-SNE Dimension 1")
    plt.ylabel("t-SNE Dimension 2")
    plt.colorbar(label="Time Index")
    plt.show()
```

```python
for metric in ['euclidean', 'cosine', 'manhattan']:
    tsne = TSNE(n_components=2, metric=metric, random_state=42)
    data_2d = tsne.fit_transform(neural_data)

    plt.figure(figsize=(8, 6))
    plt.scatter(data_2d[:, 0], data_2d[:, 1], c=np.arange(len(data_2d)), cmap=plt.cm.Spectral)
    plt.title(f"t-SNE with Metric={metric}")
    plt.xlabel("t-SNE Dimension 1")
    plt.ylabel("t-SNE Dimension 2")
    plt.colorbar(label="Time Index")
    plt.show()
```

```python
# Add more distinct patterns to the data
for i in range(0, n_time_steps, 250):
    if i % 500 == 0:
        neural_data[i:i+100, :] += np.cos(np.linspace(0, np.pi, n_neurons))  # Add cosine patterns
    else:
        neural_data[i:i+100, :] += np.exp(-np.linspace(0, 1, n_neurons))  # Add exponential patterns
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
data_2d = tsne.fit_transform(neural_data)

# Plot t-SNE results
plt.figure(figsize=(8, 6))
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=np.arange(len(data_2d)), cmap=plt.cm.Spectral)
plt.title("t-SNE Visualization of Neural Data with Structured Patterns")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.colorbar(label="Time Index")
plt.show()
```

```python
# Assign labels
labels = np.zeros(n_time_steps)
labels[0:100] = 1  # Cosine-dominant
labels[500:600] = 2  # Exponential-dominant
labels[250:300] = 3  # Transition region
labels[850:] = 4     # Another transition region

# Plot with labels
plt.figure(figsize=(8, 6))
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=labels, cmap=plt.cm.tab10)
plt.title("t-SNE Visualization with Cluster Labels")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.colorbar(label="Cluster Label")
plt.show()
```

```python
for perplexity in [5, 30, 50]:
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    data_2d = tsne.fit_transform(neural_data)

    plt.figure(figsize=(8, 6))
    plt.scatter(data_2d[:, 0], data_2d[:, 1], c=np.arange(len(data_2d)), cmap=plt.cm.Spectral)
    plt.title(f"t-SNE with Perplexity={perplexity}")
    plt.xlabel("t-SNE Dimension 1")
    plt.ylabel("t-SNE Dimension 2")
    plt.colorbar(label="Time Index")
    plt.show()
```
