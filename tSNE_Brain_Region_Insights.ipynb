```python
from google.colab import drive
drive.mount('/content/drive')
import os

# Path to your directory in Google Drive
drive_path = '/content/drive/My Drive/mini_projects/eeg_dataset'

# List all CSV files in the specified directory
files = [f for f in os.listdir(drive_path) if f.endswith('.csv')]
print("CSV Files in the Directory:", files)
```

```python
import pandas as pd

# Create a dictionary of DataFrames
dataframes = {}
for file in files:
    filepath = os.path.join(drive_path, file)
    dataframes[file] = pd.read_csv(filepath)
    print(f"Loaded {file} with shape: {dataframes[file].shape}")
```

```python
# Inspect the first file
first_file = files[0]
print(f"Inspecting file: {first_file}")
print(dataframes[first_file].head())
```

```python
for file, df in dataframes.items():
    print(f"{file}: Missing Values")
    print(df.isnull().sum())
```

```python
# Rename columns for clarity
for file in dataframes:
    dataframes[file].columns = [f'Channel_{i+1}' for i in range(dataframes[file].shape[1])]

# Verify the renaming
print(dataframes['s28.csv'].head())

# Verify the renaming
print(dataframes['s29.csv'].head())
```

```python
combined_df = pd.concat(
    [df.assign(Subject=file) for file, df in dataframes.items()],
    ignore_index=True
)
print("Combined DataFrame Shape:", combined_df.shape)
print(combined_df.head())
```

```python
# Normalize all channels (exclude 'Subject' column)
channels = combined_df.columns.difference(['Subject'])
combined_df[channels] = (combined_df[channels] - combined_df[channels].mean()) / combined_df[channels].std()

# Verify normalization
print(combined_df.head())
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Select numerical data for t-SNE (exclude 'Subject' column)
sampled_df = combined_df.sample(n=10000, random_state=42)
features_sampled = sampled_df.drop(columns=['Subject'])

# Apply t-SNE to the sampled data
tsne = TSNE(n_components=2, perplexity=10, random_state=42)
data_2d = tsne.fit_transform(features_sampled)

# Visualize t-SNE results
plt.figure(figsize=(8, 6))
plt.scatter(
    data_2d[:, 0],
    data_2d[:, 1],
    c=sampled_df['Subject'].astype('category').cat.codes,  # Color by Subject for sampled data
    cmap='viridis',
    alpha=0.7
)
plt.title("t-SNE Visualization of EEG Data")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.colorbar(label="Subject ID")
plt.show()
```

```python
# Apply t-SNE to the sampled data
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
data_2d = tsne.fit_transform(features_sampled)

# Visualize t-SNE results
plt.figure(figsize=(8, 6))
plt.scatter(
    data_2d[:, 0],
    data_2d[:, 1],
    c=sampled_df['Subject'].astype('category').cat.codes,  # Color by Subject for sampled data
    cmap='viridis',
    alpha=0.7
)
plt.title("t-SNE Visualization of EEG Data")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.colorbar(label="Subject ID")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(data_2d)

# Visualize clusters
plt.figure(figsize=(8, 6))
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=clusters, cmap='tab10', alpha=0.7)
plt.title("t-SNE with K-Means Clustering")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.colorbar(label="Cluster ID")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Perform K-Means clustering on t-SNE-transformed data
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(data_2d)

# Add clusters to the sampled DataFrame
sampled_df['Cluster'] = clusters

# Count the number of points from each subject in each cluster
cluster_subject_distribution = sampled_df.groupby(['Cluster', 'Subject']).size().unstack(fill_value=0)
print(cluster_subject_distribution)

# Visualize cluster contribution by subject
cluster_subject_distribution.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.title("Subject Contribution to Clusters")
plt.xlabel("Cluster ID")
plt.ylabel("Number of Points")
plt.legend(title="Subject", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Ensure only numeric columns are used for variance calculations
numeric_columns = sampled_df.select_dtypes(include=['float64', 'int64']).columns
numeric_data = sampled_df[numeric_columns]

# Compute variance for each channel within each cluster
channel_variances_by_cluster = numeric_data.groupby(sampled_df['Cluster']).var()

# Identify top high-variance channels per cluster
top_channels_per_cluster = channel_variances_by_cluster.idxmax(axis=1)
print("Top High-Variance Channels for Each Cluster:")
print(top_channels_per_cluster)

# Plot heatmap of variances
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(channel_variances_by_cluster, cmap="YlGnBu", annot=True, fmt=".2f")
plt.title("Channel Variances Across Clusters")
plt.xlabel("Channels")
plt.ylabel("Clusters")
plt.show()

# Check subject contribution in clusters with high variance channels
for cluster, top_channel in top_channels_per_cluster.items():
    print(f"Cluster {cluster}: Top Channel {top_channel}")
    subject_contribution = sampled_df[sampled_df['Cluster'] == cluster]['Subject'].value_counts()
    print(f"Subjects in Cluster {cluster}:")
    print(subject_contribution)
```

```python
# Define the mapping of EEG channels to brain regions
channel_brain_region_map = {
    'Fp1': 'Prefrontal Cortex', 'Fp2': 'Prefrontal Cortex',
    'F3': 'Frontal Cortex', 'F4': 'Frontal Cortex', 'F7': 'Frontal Cortex',
    'F8': 'Frontal Cortex', 'Fz': 'Frontal Cortex',
    'T3': 'Temporal Cortex', 'T4': 'Temporal Cortex',
    'T5': 'Temporal Cortex', 'T6': 'Temporal Cortex',
    'C3': 'Central Cortex', 'C4': 'Central Cortex', 'Cz': 'Central Cortex',
    'P3': 'Parietal Cortex', 'P4': 'Parietal Cortex', 'Pz': 'Parietal Cortex',
    'O1': 'Occipital Cortex', 'O2': 'Occipital Cortex'
}

# Print the mapping for reference
print("Channel to Brain Region Mapping:")
for channel, region in channel_brain_region_map.items():
    print(f"{channel}: {region}")

# Map existing columns in the dataset (Channel_1 to Channel_19) to the brain regions
# Assuming 'Channel_1', ..., 'Channel_19' correspond to the EEG channels in this order:
eeg_channel_order = ['Fp1', 'Fp2', 'F3', 'F4', 'F7', 'F8', 'T3', 'T4', 'C3', 'C4',
                     'T5', 'T6', 'P3', 'P4', 'O1', 'O2', 'Fz', 'Cz', 'Pz']

# Map Channel numbers to brain regions
channel_region_mapping = {
    f"Channel_{i+1}": channel_brain_region_map[eeg_channel_order[i]]
    for i in range(len(eeg_channel_order))
}

# Print the mapping for reference
print("\nMapped Channels to Brain Regions:")
for channel, region in channel_region_mapping.items():
    print(f"{channel}: {region}")
```

```python
# Compute variance for each channel within each cluster, excluding 'Subject'
numerical_columns = sampled_df.drop(columns=['Subject']).columns  # Exclude 'Subject'
channel_variances_by_cluster = sampled_df.groupby('Cluster')[numerical_columns].var()

# Identify the high-variance channels for each cluster
top_channels_per_cluster = channel_variances_by_cluster.idxmax(axis=1)
print("Top High-Variance Channels for Each Cluster:")
print(top_channels_per_cluster)

# Map the top high-variance channels to brain regions
print("\nTop High-Variance Channels and Their Brain Regions:")
for cluster, top_channel in top_channels_per_cluster.items():
    brain_region = channel_region_mapping[top_channel]
    print(f"Cluster {cluster}: {top_channel} ({brain_region})")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot heatmap of variances
plt.figure(figsize=(12, 6))
sns.heatmap(channel_variances_by_cluster, cmap="YlGnBu", annot=True, fmt=".2f")
plt.title("Channel Variances Across Clusters")
plt.xlabel("Channels")
plt.ylabel("Clusters")
plt.show()

# Count subject contributions to each cluster
subject_contribution_by_cluster = sampled_df.groupby('Cluster')['Subject'].value_counts()

# Display contributions for clusters with the highest-variance channels
for cluster, top_channel in top_channels_per_cluster.items():
    print(f"\nCluster {cluster} (Top Channel: {top_channel}):")
    print(subject_contribution_by_cluster.loc[cluster])
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Brain region mapping
channel_to_brain_region = {
    'Channel_1': 'Frontal Cortex', 'Channel_2': 'Prefrontal Cortex', 'Channel_3': 'Frontal Cortex',
    'Channel_4': 'Frontal Cortex', 'Channel_5': 'Frontal Cortex', 'Channel_6': 'Frontal Cortex',
    'Channel_7': 'Temporal Cortex', 'Channel_8': 'Temporal Cortex', 'Channel_9': 'Parietal Cortex',
    'Channel_10': 'Parietal Cortex', 'Channel_11': 'Occipital Cortex', 'Channel_12': 'Occipital Cortex',
    'Channel_13': 'Central Cortex', 'Channel_14': 'Central Cortex', 'Channel_15': 'Central Cortex',
    'Channel_16': 'Occipital Cortex', 'Channel_17': 'Parietal Cortex', 'Channel_18': 'Temporal Cortex',
    'Channel_19': 'Temporal Cortex'
}

# Cluster-wise top channels
top_channels = {
    0: 'Channel_16',
    1: 'Channel_6',
    2: 'Channel_5',
    3: 'Channel_2',
    4: 'Channel_5'
}

# Map top channels to brain regions
top_regions = [channel_to_brain_region[channel] for channel in top_channels.values()]

# Count occurrences of each brain region
region_counts = Counter(top_regions)

# Plot brain region distribution
plt.figure(figsize=(8, 6))
plt.bar(region_counts.keys(), region_counts.values(), color='skyblue')
plt.title("Brain Region Distribution Across Clusters")
plt.xlabel("Brain Region")
plt.ylabel("Number of Clusters")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Plot top variance channels by cluster
plt.figure(figsize=(8, 6))
plt.bar(top_channels.keys(), range(len(top_channels)), color='lightgreen', tick_label=list(top_channels.values()))
plt.title("Top Variance Channels by Cluster")
plt.xlabel("Cluster")
plt.ylabel("Channel")
plt.tight_layout()
plt.show()
```
